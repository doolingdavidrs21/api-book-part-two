Whether you called this API from the web browser or from Python, you executed an HTTP GET verb, which is used for reading information from an API. If you wanted to call a REST API to take other actions, you might use HTTP verbs such as POST, PUT, or DELETE. You will notice that REST APIs have separate endpoints for each action. For instance, this API would have separate endpoints to get a list of players or teams. The API defines the fields in the response, and they are the same for each request.

Another architectural style that is useful for data science is GraphQL, which has several differences from REST. Where REST APIs have multiple endpoints, a GraphQL API has only one. Where REST APIs return the same fields each time an endpoint is called, GraphQL allows the consumer to define what fields they want. Where REST uses an HTTP GET when reading data, GraphQL uses an HTTP POST. Since a web browser by default sends an HTTP GET, you can’t directly call a GraphQL API in the browser. However, some GraphQL APIs provide a web interface that allows you to query the API.

To make a request to a GraphQL API in Python, you can send a query in a POST command using the httpx library as follows:

````python
import httpx

graphql_url = "https://countries.trevorblades.com"

json_query = {'query': '''
{
  country(code: "US") {
    name
    native
    currency
    languages {
      code
      name
      }
    }
}
'''}

api_response = httpx.post(graphql_url, json=json_query)
print(api_response.json())
````

### The Problem

`uv` found a package named `setuptools` (or `fastapi`) on **TestPyPI** and stopped looking elsewhere. It assumed that if the package exists on the index you specified, you must want *that* version. Unfortunately, TestPyPI is full of incomplete or broken uploads (often by bots or students), so `uv` tried to build a broken version and failed.

### The Fix

You need to explicitly tell `uv` allowed to compare versions across **both** indexes (Real PyPI and TestPyPI) and pick the one that actually works.

Run this command instead:

```bash
uv add swcpy-doolingdavidrs21 \
  --extra-index-url https://test.pypi.org/simple/ \
  --index-strategy unsafe-best-match
```

### What `--index-strategy unsafe-best-match` does

It tells `uv`: "I trust both of these sources. If you find a package on TestPyPI that is broken or doesn't match my requirements, **keep looking** on the real PyPI."

*Note: The warning `Indexes specified via ... will not be persisted` means that if you run `uv sync` tomorrow without these flags, it might complain about missing `swcpy-doolingdavidrs21`. But for this tutorial step, this command is exactly what you need.*

This is the classic **"TestPyPI Junkyard Problem."** I completely understand the frustration; this specific error is nasty because it looks like a compilation issue, but it's actually a dependency resolution issue.

### The Problem: "Newer" isn't always "Better"

1.  **The Trap:** Someone (likely a bot or a student) uploaded a broken, half-baked version of `pyarrow` (version 22.0.0) to **TestPyPI**.
2.  **The Mistake:** Real PyPI only has version `18.x`.
3.  **The Crash:** When you ran `--index-strategy unsafe-best-match`, `uv` looked at both indexes, saw version `22.0.0` on TestPyPI, and said, *"Wow\! Version 22 is newer than 18\! I'll take that one\!"*
4.  **The Explosion:** That version `22.0.0` is broken (it has no pre-built binaries), so your computer tried to compile `pyarrow` from raw C++ code and failed.

### The Fix: Force the "Real" Version

You need to tell `uv`: "Install my SDK from TestPyPI, but **force** pyarrow to stay on a version that actually exists on the real PyPI."

Run this command exactly:

```bash
uv add swcpy-doolingdavidrs21 "pyarrow<20" \
  --extra-index-url https://test.pypi.org/simple/ \
  --index-strategy unsafe-best-match
```

### Why this works

By adding `"pyarrow<20"`, you are effectively banning the broken version `22.0.0` found on TestPyPI. `uv` will be forced to ignore the "junk" version and download the stable, pre-compiled wheel (approx version 18.x) from the real PyPI.
